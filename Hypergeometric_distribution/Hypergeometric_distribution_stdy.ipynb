```python
import matplotlib.pyplot as plt
from scipy.stats import hypergeom
import numpy as np
```

```python
# Fixed total population size
N = 100

# Try varying m (successes in population)
m_values = [i for i in range(0, 101, 10)]
k = 20  # sample size

for m in m_values:
    x = np.arange(0, k + 1)  # possible values of x (successes in sample)
    rv = hypergeom(N, m, k)
    probs = rv.pmf(x)

    plt.plot(x, probs, label=f"m = {m}")

plt.title(f"Hypergeometric PMF (N={N}, K={k})")
plt.xlabel("x (successes in sample)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fixed total population size
N = 100

k = 90  # sample size
# Try varying m (successes in population)
m_values = [i for i in range(k, N+1, 1)]

for m in m_values:
    x = np.arange(77, k + 1)  # possible values of x (successes in sample)
    rv = hypergeom(N, m, k)
    probs = rv.pmf(x)
    print(sum(probs))

    plt.plot(x, probs, label=f"m = {m}")

plt.title(f"Hypergeometric PMF (N={N}, K={k})")
plt.xlabel("x (successes in sample)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fixed total population size
N = 100

k = 90  # sample size
# Try varying m (successes in population)
m_values = [i for i in range(k, N+1, 1)]

for m in m_values:
    x = np.arange(77, k + 1)  # possible values of x (successes in sample)
    rv = hypergeom(N, m, k)
    probs = rv.pmf(x)
    print(sum(probs))

    plt.plot(x, probs, label=f"og rule = {m}")

plt.title(f"Hypergeometric PMF (pruned rule rec={N}, pruned rule fails={k})")
plt.xlabel("x (fails in sample)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```
